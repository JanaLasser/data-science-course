```python
url = 'http://download-data.deutschebahn.com/static/datasets/sensordaten_schenker/161209_Schenker_Sensordaten.csv'
```

```python
from urllib import request
response = request.urlopen(url)
```

```python
data = response.read()
```

```python
st = data.decode('utf-8')
```

```python
import pandas as pd
import io
buf = io.StringIO(st)
```

```python
df = pd.read_csv(buf)
```

```python
set(df['SXXJ number'])
```

```python
pos = df[df['SXXJ number']=='A'][['Latitude', 'Longitude']]
```

```python
pos = pos[pd.notnull(pos['Latitude']) & pd.notnull(pos['Longitude'])]
```

```python
pos = pos.astype(float)
```
